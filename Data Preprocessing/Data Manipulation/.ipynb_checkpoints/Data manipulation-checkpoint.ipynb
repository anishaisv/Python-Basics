## DATA MANIPULATION ##


#### Data manipulation means changing data so it's easier to understand or use.

###### Remove junk → Delete wrong or extra data
###### Sort stuff → Arrange data by date, name, or number
###### Group things → Put similar data together (like grouping clothes by color)
###### Add or fix things → Fill in missing info or create new columns
###### Combine boxes → Join data from different places into one table
