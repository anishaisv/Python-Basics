## DATA CLEANING

```python
#loading the dataset 
import pandas as pd
df = pd.read_csv("dc_dataset.csv")
df.head()
```

```python
df.rename(columns={"No_Pation":"Patient_id_no"}) # Renaming 2nd column
```

#### Checking Duplicates

```python
df.duplicated().sum() # result is 0, so no duplicates
```

#### Check missing values

```python
df.isna().sum()
```

#### Fill using mean

```python
hmean=round(df['HbA1c'].mean(),2) #calculate mean for HbA1c
hmean
df['HbA1c'].fillna(hmean,inplace=True)
df
```

```python
df.isna().sum()#you can see no missing values in HbA1c field
```

```python
#Since its medical dataset we can't just take the mean for all bz patient's information are sensitive.so we will dropp all missing values
finaldata=df.dropna()
finaldata
```

```python
finaldata.isna().sum() # see there are no missing values now.
```

### Cleaning OUTLIERS
#### Using Boxplot

```python
# boxplot helps you spot outliers by showing you where most of your data sits
import seaborn as sns
sns.boxplot(finaldata["Urea"]) # use seaborn library as sns; The circle corresponding to 40 is an outlier. 
```

```python
sns.boxplot(finaldata["Cr"]) # here value near 800 is an outlier.
```

```python
maxol=finaldata["Cr"].quantile(0.99) #Let's find the outlier; find maximum value first. which will be in the last quartile
maxol
```

```python
#no. 327 is not the value. Increase the quantile value
maxol=finaldata["Cr"].quantile(0.995)
maxol
```

```python
#Now we found the max value as per boxplot diagram. any value above 401 is an outlier. 
# Find all the values above 401 in the column Cr
finaldata["Cr"]>maxol  
```

```python
                                # we are getting only boolean value inorder to get the records,pass it to a dataframe again
finaldata[finaldata["Cr"]>maxol]  # important thing to find the records,pass it to a dataframe again
                                    # from the result, under column 'Cr' u can see 4 records are holding the value 800(outliers)
```

```python
# Now we are gonna ignore these outliers. so keep all values <= 401 in the final dataset
finaldataset=finaldata[finaldata["Cr"]<=maxol]
finaldataset
```

```python
# Let's plot the bloxplot now to check the outliers.
sns.boxplot(finaldataset["Cr"])
```
