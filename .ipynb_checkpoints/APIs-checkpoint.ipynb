```python
import requests # when communicating with a website(source)- first thing - check for response.  for that inmport request - requesting data from website
```

```python
url='http://numbersapi.com/42?json'  #json file- similar to dictionary, key value pair data
```

```python
response=requests.get(url) #to check whether the url is active or not. if response is 404 not active, 200 is active
```

```python
response   # 200- which means website is still working
```

```python
response.text # to check what data available in that website
```

```python
print(response.text)  # for clear understanding
```
