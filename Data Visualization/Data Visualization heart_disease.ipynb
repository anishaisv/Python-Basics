```python
# Simple Plot
# Provide Labels to the axis, Title on the Graph
from matplotlib import pyplot as plt
x = [0,1,2,3,4]
y = [0,10,20,30,40]
plt.plot(x,y)
plt.title("Simple Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
#plt.show()
```

```python
# Highlight the intersection points. using MARKER
from matplotlib import pyplot as plt
x = [0,1,2,3,4]
y = [0,10,20,30,40]
plt.plot(x,y)
plt.plot(x,y, marker="o")
plt.title("Simple Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# Highlight the intersection points. using MARKER
from matplotlib import pyplot as plt
x = [0,1,2,3,4]
y = [0,10,20,30,40]
plt.plot(x,y)
plt.plot(x,y, marker="h",ms=15) # increase markersize
plt.title("Simple Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# Style the Plot; font style, size,color

from matplotlib import pyplot as plt
x = [0,1,2,3,4]
y = [0,10,20,30,40]
plt.plot(x,y)
plt.plot(x,y, marker="o",c="purple")
#font families
# Create font family and title using dictionary
f1= {"family":"calibri","color":"red","size":"25"}
f2 = {"family":"calibri","color":"blue","size":"25"}

plt.title("Simple Plot",fontdict=f1)             # fontdict is the parameter used to exhibit all in f1
plt.xlabel("X-Axis",fontdict=f2)
plt.ylabel("Y-Axis",fontdict=f2)
plt.show() 
```

## Sub-Plots
#### Using same space; but using multiple plots in that representation

```python
from matplotlib import pyplot as plt

x = [0,1,2,3,4]
y = [0,10,20,30,40]
y1 = [30,20,40,10,50]
plt.plot(x,y,x,y1) #x,y -horizontal and vertical axis
plt.title("Sub Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
 
```

```python
from matplotlib import pyplot as plt

x = [0,1,2,3,4]
y = [0,10,20,30,40]
y1 = [30,20,40,10,50]
plt.subplot(1,2,1) # 1- 2- dividing into 2 parts 1-this is 1st plot
plt.plot(x,y)
plt.title("Sub Plot 1")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.subplot(1,2,2)
plt.plot(x,y1)
plt.title("Sub Plot 2")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
from matplotlib import pyplot as plt

x = [0,1,2,3,4]
y = [0,10,20,30,40]
y1 = [30,20,40,10,50]
plt.subplot(2,1,1) # 1- ; 2- dividing into 2 parts; 1-this is 1st plot
plt.plot(x,y)
plt.title("Sub Plot 1")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.subplot(2,1,2)
plt.plot(x,y1)
plt.title("Sub Plot 2")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

## HEART DISEASE DATASET

```python
#Working on the dataset
import pandas as pd
df = pd.read_csv("heart.csv")
df.head()
```

## Heat Map

```python
import seaborn as sns

# Visualize missing values in the DataFrame.
# Each cell represents whether the value is missing (True) or not (False).
# This helps identify patterns or concentrations of missing data.
sns.heatmap(df.isnull())
```

## Histogram

```python
import matplotlib.pyplot as plt
## Histogram; # Plot a histogram of the 'age' column to visualize the distribution of age values in the dataset.
plt.hist(df['age'])
```

```python
import matplotlib.pyplot as plt
## Histogram; # Plot a histogram of the 'age' column to visualize the distribution of age values in the dataset.
plt.hist(df['age'])
```

```python
# Plot a histogram of the 'age' column with custom bin ranges and black edges.
# This groups ages into the intervals: 20–40, 40–60, and 60–90.
plt.hist(df['age'], edgecolor='black', bins=[20, 40, 60, 90])
```

## Boxplots

```python
# Box plot of the 'age' column to visualize its distribution, median, and potential outliers.
# Given values: 22, 22, 27, 33, 38, 40, 60, 62, 66, 69
# Median appears to be around 40, showing the central tendency of the data.
#22, 22, 27, 33, 38, 
#40  - median
#40, 60, 62, 66, 69
plt.boxplot(df['age'])
```

## Scatterplot

```python
# Scatter plot to explore the relationship between age and resting blood pressure (trestbps).
# Each point represents an individual's age and their corresponding resting blood pressure.
plt.scatter(df['age'], df['trestbps'])
```

## Jointplot

```python
# Joint plot to visualize the relationship between age and resting blood pressure (trestbps),
# along with their individual distributions on the axes.
sns.jointplot(x=df['age'], y=df['trestbps'], data=df)
```
